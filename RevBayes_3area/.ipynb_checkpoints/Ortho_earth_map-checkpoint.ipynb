```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


m = Basemap(projection='ortho',
            resolution='c',
            lat_0=90,
            lon_0=90)

# fill background.
m.drawmapboundary(fill_color='#deebf7')

# draw coasts and fill continents.
#m.drawcoastlines(linewidth=0.2)
m.fillcontinents(color='#31a354',lake_color='#deebf7')

# 20 degree graticule.
m.drawparallels(np.arange(-80,81,10), color='grey', linewidth=0.5)
m.drawmeridians(np.arange(-180,180,10), color='grey', linewidth=0.5)

#plt.show()

plt.savefig('Ortho_earth_map.tif', dpi=300, bbox_inches="tight", pad_inches=0)
```
